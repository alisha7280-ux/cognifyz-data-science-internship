## Task 3: Geospatial Analysis


```python
df[['Latitude', 'Longitude']].isnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.3)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Restaurants')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Longitude'], df['Latitude'], c=df['Aggregate rating'], alpha=0.5)
plt.colorbar(label='Aggregate Rating')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Restaurant Locations Colored by Rating')
plt.show()
```

```python
df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('Country Code')['Aggregate rating'].mean().sort_values(ascending=False).head()
```

### Task 3 Conclusion

Geospatial analysis was performed using latitude and longitude to visualize the distribution of restaurants. The analysis showed that restaurants are concentrated in specific geographic regions. By mapping aggregate ratings to locations, variations in restaurant ratings across different cities and regions were observed.
