## Level 2: Data Analysis Tasks

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r"C:\Users\DELL\Downloads\Dataset.xlsx")  # use correct path
```

## Task 1: Table Booking & Online Delivery


```python
df['Has Table booking'].value_counts(normalize=True) * 100
```

```python
df['Has Online delivery'].value_counts(normalize=True) * 100
```

```python
df.groupby('Has Table booking')['Aggregate rating'].mean()
```

```python
df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True)
```

```python
df.groupby('Has Table booking')['Aggregate rating'].mean().plot(kind='bar')
plt.ylabel('Average Rating')
plt.title('Average Rating by Table Booking Availability')
plt.show()
```

### Level 2 – Task 1 Conclusion

The percentage of restaurants offering table booking and online delivery was calculated. Restaurants with table booking were observed to have slightly higher average ratings compared to those without. Online delivery availability varied across different price ranges.


## Level 2 – Task 2: Price Range Analysis


```python
df['Price range'].value_counts()
```

```python
df.groupby('Price range')['Aggregate rating'].mean()
```

```python
df.groupby('Price range')['Aggregate rating'].mean().plot(kind='bar')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.title('Average Rating by Price Range')
plt.show()
```

```python
df.groupby('Price range')['Aggregate rating'].mean().idxmax()
```

### Level 2 – Task 2 Conclusion

The most common price range among restaurants was identified. The average aggregate rating was calculated for each price range, and it was observed that higher price ranges generally tend to have higher average ratings.


## Level 2 – Task 3: Feature Engineering


```python
df['Table_Booking_Flag'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})

```

```python
df['Online_Delivery_Flag'] = df['Has Online delivery'].map({'Yes': 1, 'No': 0})
```

```python
df['Restaurant_Name_Length'] = df['Restaurant Name'].str.len()
```

```python
df['Address_Length'] = df['Address'].str.len()
```

```python
df[['Table_Booking_Flag', 'Online_Delivery_Flag', 'Restaurant_Name_Length', 'Address_Length']].head()
```

### Level 2 – Task 3 Conclusion

New features were created from existing columns to support further analysis and modeling. Binary variables were generated for table booking and online delivery, and text-based features such as restaurant name and address length were extracted.
